1. What does an empty dictionary's code look like?

Ans: Empty dictionary looks like a set. {}

2. What is the value of a dictionary value with the key 'foo' and the value 42?

Ans : The value is 42. {"foo":42}

3. What is the most significant distinction between a dictionary and a list?

Ans: Dictionary has key, value pairs whereas list has only elements. There are no pairs in a list
List has indexing property whereas there is no such indexing property in a dictionary, in dictionary we fetch a value by its key and not by its index number
Items in a List are ordered however there is no such order in dictionary.

4. What happens if you try to access spam['foo'] if spam is {'bar': 100}?

Ans: We will get KeyError as there is no key named 'foo' in the dictionary mentioned in the question

5. If a dictionary is stored in spam, what is the difference between the expressions 'cat' in spam and 'cat' in spam.keys()?

Ans: There is no difference. In both the cases the computer will search for 'cat' in the keys of spam.

6. If a dictionary is stored in spam, what is the difference between the expressions 'cat' in spam and 'cat' in spam.values()?

Ans: 'cat' in spam will search for a key named 'cat' whereas in 'cat' in spam.values(), the computer will search for 'cat' in values

7. What is a shortcut for the following code?
if 'color' not in spam:
spam['color'] = 'black'

Ans: spam.setdefault('color','black')
